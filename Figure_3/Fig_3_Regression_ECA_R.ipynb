## Analysis with data pooled together of 5 days

```r
Cellulose <- read.csv('Rate_Sub_Enz_5_Day.csv')
head(Cellulose)
```

```r
Rate <- Cellulose$Rate
Substrate <- Cellulose$Substrate
Enzyme <- Cellulose$Enzyme
```

```r
#the formula for the model
log_formula <- formula(Rate ~ (Substrate*Enzyme)*Vmax/(K + Substrate + Enzyme))

#fit the model
m <- nls(log_formula,start=list(Vmax= 1,K=100))

#estimated parameters
summary(m)
```

```r
plot(Substrate,Rate)
lines(Substrate,predict(m),col="red",lty=2,lwd=3)
```
