```r
extract_data <- function(filename){

    load(filename,ex <- new.env()) 
    ls.str(ex)
    out <- ex$out[[1]]
    #[,3] refers to cellulose
    Substrate_series <- out$SubstratesSeries[-1,3]
    
#     Substrate_average <- NULL
#     for (i in 1:6){
#         Substrate_average <-c(Substrate_average,mean(Substrate_series[1:50*(i)]))
#     }
#     Substrate_average <- Substrate_average[1:6]
    

    Enzyme_Sub <- out$Enzyme_subSeries
    colnames <- rownames(Enzyme_Sub)[1:12]
    Enzyme_Sub1 <- matrix(t(Enzyme_Sub),ncol=12,byrow=T)
    colnames(Enzyme_Sub1) <- colnames
    Enzyme_series <- Enzyme_Sub1[-1,3]
    
#     Enzyme_average <- NULL
#     for (i in 1:6){
#         Enzyme_average <-c(Enzyme_average,mean(Enzyme_series[1:50*(i)]))
#     }
#     Enzyme_average <- Enzyme_average[1:6]

    
    
    decayrate <- out$DecayRatesSeries
    colnames <- rownames(decayrate)[1:12]
    decayrate1 <- matrix(t(decayrate),ncol=12,byrow=T)
    colnames(decayrate1) <- colnames

    Decayrates_series <- decayrate1[-1,3]
    
#     Decayrates_average <- NULL
#     for (i in 1:6){
#         Decayrates_average <-c(Decayrates_average,mean(Decayrates_series[1:50*(i)]))
#     }
#     Decayrates_average <- Decayrates_average[1:6]
    
    cellulose <- cbind(Substrate_series,Enzyme_series,Decayrates_series)}

```

```r
datalist = lapply(filelist, function(x)extract_data(x)) 
```
